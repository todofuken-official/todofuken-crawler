```python
import requests
from bs4 import BeautifulSoup

response = requests.get("https://www.walkerplus.com/event_list")
response.encoding = 'utf-8'
soup = BeautifulSoup(response.text, "html.parser")

events = soup.select(".m-mainlist__item")

for event in events: 
    title_tag = event.select_one(".m-mainlist-item__ttl > span")
    if not title_tag:
        continue
    title = title_tag.text.strip()

    period_tag = event.select_one(".m-mainlist-item-event__period")
    if period_tag:
        open_tag = period_tag.select_one(".m-mainlist-item-event__open")
        if open_tag:
            open_tag.decompose()
        period = period_tag.text.strip()
    else:
        period = ""

    desc = event.select_one(".m-mainlist-item__txt").text

    areas = event.select(".m-mainlist-item__map > a")
    area_text = " ".join([area.text.strip() for area in areas])

    station_tags = event.select(".m-mainlist-item__station > a")
    stations = [station.text.strip() for station in station_tags]
    stations_text = ", ".join(stations)

    place = event.select_one(".m-mainlist-item-event__place").text

    tag_links = event.select(".m-mainlist-item__tagsitemlink")
    tags = [tag.text.strip() for tag in tag_links]
    tags_text = ", ".join(tags)

    print(title, period, desc, area_text, stations_text, place, tags_text)
```
